# **Reading Data From Json**

```python
import json
cd=open("products.json","r")
rd=cd.read()
cd.close()

products=json.loads(rd)
```

```python
products
```

# **Adding Items To Inventory**

```python
id=int(input("Enter the product id: ")) #asking for new item to be added in inventory
name=input("Enter the name of the product: ")
weight=int(input("Enter the net weight of the product: "))
price=int(input("Enter the price of the product:  "))
quantity=int(input("Enter the quantity of the product:  "))
size=input("Enter the size of the product: ")

products[id]= {"Name":name,"Weight":weight,"Price":price,"Quantity":quantity,"Size":size} #for adding new item
products #new item have been added
```

# **User Enters Another Quantity If Item Exists**

```python
id=int(input("Enter the product id: ")) #asking for new item to be added in inventory
name=input("Enter the name of the product: ")
weight=int(input("Enter the net weight of the product: "))
price=int(input("Enter the price of the product:  "))
quantity=int(input("Enter the quantity: "))
size=input("Enter the size of the product: ")

products[id]= {"Name":name,"Weight":weight,"Price":price,"Quantity":quantity,"Size":size} #for adding new item
quant=int(input("Enter the new quantity you want to add in the current item: "))
print("The final quantity is:",products[id]['Quantity']+ quant)
products[id]['Quantity']=products[id]['Quantity']+quant
```

```python
products #finally the total quantity has been added to the inventory
```

## **Updating Inventory After Adding Quantity**

```python
ab=json.dumps(products)

cd=open("products.json","w")
cd.write(ab)  
cd.close()
```
