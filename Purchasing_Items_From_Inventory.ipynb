# **Reading Data From Inventory**

```python
import json
cd=open("products.json","r")
rd=cd.read()
cd.close()

products=json.loads(rd)
```

```python
products
```

# **Purchasing Items From Inventory**

```python
pur_id=input("Enter the product id: ") #Billing amount of first item
pur_quant=int(input("Enter the quantity: "))

print("Product: " ,products[pur_id]['Name'])
print("Price:  ",products[pur_id]['Price'])
print("Your billing amount is: ",products[pur_id]['Price'] * pur_quant)

products[pur_id]['Quantity']=products[pur_id]['Quantity'] - pur_quant
```

# **Updating Inventory After Purchase**

```python
ab=json.dumps(products)

cd=open("products.json","w")
cd.write(ab)  
cd.close()
```
